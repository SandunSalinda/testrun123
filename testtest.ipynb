```python



























































import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
img = cv.imread('j.png', cv.IMREAD_GRAYSCALE)

 # Display the original image using matplotlib
2 plt.imshow(img_noisy1, cmap='gray')
3 plt.title('Original Image')
4 plt.axis('off') # Hide axes for a cleaner look
5 plt.show()
6
7
8 # Display the blurred image using matplotlib
9 plt.imshow(img_new1, cmap='gray')
10 plt.title('Median Filtered Image')
11 plt.axis('off') # Hide axes for a cleaner look
12 plt.show(
import matplotlib.pyplot as plt
# Add Gaussian noise to the image   


 import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
[] # Load the image
E
image = cv2.imread("mage-with-bad-contrast_Q320.jpg", cv2.IMREAD_GRAYSCALE)
#Calculate the histogram
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
#[image] The image we want the histogram of (must be in a list).
#[0] The index of the channel to use. Grayscale has only 1 channel → index 0.
#None No mask is applied use the whole image.
#[256] Number of bins (groups of pixel values). We want o to 255 256 bins.
#[0, 256] The pixel value range we are analyzing (from 0 to 255).
# Display the original image
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
# Display the histogram
plt.subplot(1, 2, 2)
plt.plot(histogram)
plt.title('Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
# Show the plots
plt.tight_layout()
plt.show()


# Load the image
image = cv2.imread("macaw.jpg",cv2.COLOR_RGB2BGR)
cv2_imshow(image);

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load grayscale image
image = cv2.imread("image-with-bad-contrast_Q320.jpg", cv2.IMREAD_GRAYSCALE)

# Reusable function to display image and its histogram
def show_image_and_hist(img, title):
    hist = cv2.calcHist([img], [0], None, [256], [0, 256])
    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.plot(hist, color='black')
    plt.title(f'{title} Histogram')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# Show Original
show_image_and_hist(image, 'Original Image')

# -------- Intensity Transformations -------- #

# 1. Negative Transformation
negative_img = 255 - image
show_image_and_hist(negative_img, 'Negative Transformation')

# 2. Log Transformation
c_log = 255 / np.log(1 + np.max(image))  # Scaling factor
log_img = c_log * np.log(1 + image.astype(np.float32))
log_img = np.uint8(log_img)
show_image_and_hist(log_img, 'Log Transformation')

# 3. Gamma Transformation
gamma = 2.0   # Try values like 0.4, 0.6, 1.5, 2.0
c_gamma = 255 / (np.max(image) ** gamma)
gamma_img = c_gamma * (image.astype(np.float32) ** gamma)
gamma_img = np.uint8(np.clip(gamma_img, 0, 255))
show_image_and_hist(gamma_img, f'Gamma Transformation (γ={gamma})')
```
